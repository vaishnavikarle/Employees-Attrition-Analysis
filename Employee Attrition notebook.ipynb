# Hypothesis Testing

```python
# Hypothesis 1 Overtime
# DoesOvertime vs Attrition
# Binary vs Binary
# Null hypothesis = Over time does not affect attrition
# Alternative hypothesis = It does affect employeess attrition
```

```python
from scipy.stats import chi2_contingency

# contingency table
ct = pd.crosstab(dataset['DoesOvertime'], dataset['LeftCompany'])
print('Contingency Table',ct)
```

```python
chi2, p, dof, expected = chi2_contingency(ct)
print('Chi-square value:',chi2)
print('p value:',p)
print('Degree of freedom:',dof)
print('Expected frequency:',expected)
```

```python
# Hypothesis 2 Job Satisfaction
# Job satisfaction vs Attrition
# Ordinal vs Binary(Mann Whitney U Test)
# Null hypothesis = Job satisfaction levels are for same for both who leaves and who stays
# Alternative hypothesis = Employess with lower job satisfaction are more likly to leave.
```

```python
from scipy.stats import mannwhitneyu
js_stay = dataset[dataset['LeftCompany'] == 0]['JobSatisfaction']
js_left = dataset[dataset['LeftCompany'] == 1]['JobSatisfaction']

u_stat, p_val = mannwhitneyu(js_stay, js_left, alternative = 'two-sided')
print('Mann-Whitney U statistic:', u_stat)
print('p-value:',p_val)
```

```python
# Hypothesis 3 Work life balance
# Work life balance vs Attrition
# Ordinal vs Binary(Mann Whiteney U test)
# Null hypothesis = Work life balance for both employess are same for all employess
# Alternative hypothesis = Employees with lower work life balance have more chances to attrition
```

```python
wlb_stay = dataset[dataset['LeftCompany'] == 0]['WorkLifeBalance']
wlb_leave = dataset[dataset['LeftCompany'] == 1]['WorkLifeBalance']

u_stats, p_val = mannwhitneyu(wlb_stay,wlb_leave,alternative='two-sided')
print('Mann-Whiteny U:',u_stats)
print('p-value:', p_val)
```

```python
# Hypothesis 4 appraisal rating
# Appraisal rating vs Attrition
# Ordinal vs Binary(Mann Whiteney U test)
# Null hypothesis = Appraisal rating for both employess are same for all employess
# Alternative hypothesis = Employees with lower appraisal have more chances to attrition
```

```python
ar_stay = dataset[dataset['LeftCompany'] == 0]['AppraisalRating']
ar_leave = dataset[dataset['LeftCompany'] == 1]['AppraisalRating']

u_stats, p_val = mannwhitneyu(ar_stay,ar_leave, alternative = 'two-sided')
print('Mann Whitney U:',u_stats)
print('p value:',p_val)
```
